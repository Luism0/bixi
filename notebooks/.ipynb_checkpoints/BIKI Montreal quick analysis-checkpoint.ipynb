```python
import os
import pandas as pd
```

```python
dataset2014 = pd.read_csv(r'D:\Luismo\Documents\YGROUP\OD_2014.csv', index_col='Unnamed: 0')
stations2014 = pd.read_csv(r'D:\Luismo\Documents\YGROUP\datasets_4141_8692_Stations_2014.csv')
```

## We are going to merge all data to simplify the work

```python
all2014 = pd.merge(dataset2014, stations2014, how='inner', left_on='start_station_code', right_on='code').drop(['code'], axis=1)
```

```python
all2014 = all2014.rename(columns={'name': 'start_station_name', 
                                  'latitude': 'start_station_latitude',
                                  'longitude': 'start_station_longitude'})
```

```python
all2014
```

```python
all2014 = pd.merge(all2014, stations2014, how='inner', left_on='end_station_code', right_on='code').drop(['code'], axis=1)
```

```python
all2014 = all2014.rename(columns={'name': 'end_station_name', 
                                  'latitude': 'end_station_latitude',
                                  'longitude': 'end_station_longitude'})
```

```python
all2014
```
